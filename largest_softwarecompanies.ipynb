```python
#Normalize column names:

df.columns = df.columns.str.strip().str.replace('\n',' ').str.replace('  ',' ')
df.columns = df.columns.str.replace(' ', '_').str.replace('[()]','', regex=True).str.lower()

df.columns
```

```python
#Trim string fields & remove invisible characters

str_cols = df.select_dtypes(include='object').columns
for c in str_cols:
    df[c] = df[c].astype(str).str.strip().str.replace('\xa0',' ', regex=False)

df
```

```python
#Convert numeric columns to numeric types (convert strings, commas, $ to floats for calculations)


import numpy as np

def to_numeric_strip(series, remove_chars=r'[$,()]', percent=False):
    s = series.astype(str).replace({'nan': None})
    s = s.fillna('')
    s = s.str.replace(remove_chars, '', regex=True)
    s = s.str.strip()
   
    if percent:
        s = s.str.replace('%','', regex=False)
        return pd.to_numeric(s, errors='coerce')/100
    return pd.to_numeric(s, errors='coerce')


candidate_numeric = []
for c in df.columns:
    
    samp = df[c].dropna().astype(str).sample(min(50, max(1, df[c].dropna().shape[0])))
    if any(s.str.contains(r'[\d,.$%()]').any() for s in [samp]):

        candidate_numeric.append(c)

print("candidate numeric-like columns:", candidate_numeric)


for col in ['revenue', 'market_cap', 'employees', 'profit', 'growth_rate']:
    if col in df.columns:
        if 'rate' in col or 'growth' in col or (df[col].astype(str).str.contains('%').any()):
            df[col] = to_numeric_strip(df[col], remove_chars=r'[$,()]', percent=True)
        else:
            df[col] = to_numeric_strip(df[col], remove_chars=r'[$,()]', percent=False)
```

```python
#Check & handle missing values decide whether to drop, fill, or flag mising data

df.isnull().sum().sort_values(ascending=False)

df = df.dropna(subset=['revenue_b$', 'market_cap_b$'])

for c in ['organization','headquarters']:
    if df[c].isnull().any():
        df[c] = df[c].fillna(df[c].mode().iloc[0])

df
```

```python
#Resolve inconsistent ranks (business rule) choose an rank if you had duplicate rank twice)

df = df.sort_values('revenue_b$', ascending=False).reset_index(drop=True)
df['rank_by_revenue'] = df.index + 1

df['rank'] = df['rank_by_revenue']

df
```

```python
#Standardize organization names (small spelling and case differences )

df['organization'] = df['organization'].str.strip().str.title()

df
```

```python
#Parse 'Headquarters' into city / state / country (makes geographic analysis easier)

hq_split = df['headquarters'].str.split(',', expand=True).apply(lambda x: x.str.strip())

df['hq_city'] = hq_split[0]
df['hq_state_or_region'] = hq_split[1].fillna('')
df['hq_country'] = hq_split[2].fillna(hq_split[1])

df['hq_country'] = df['hq_country'].replace({'US':'United States','U.S.':'United States','USA':'United States'})

df
```

```python
#Detect & handle outliers (option)

Q1 = df['revenue_b$'].quantile(0.25)
Q3 = df['revenue_b$'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['revenue_b$'] < lower) | (df['revenue_b$'] > upper)]
print("Outlier rows count:", outliers.shape[0])

df['revenue_outlier'] = df['revenue_b$'].apply(lambda x: x < lower or x > upper)
```

```python
#final verification before analysis

df['rank'] = pd.to_numeric(df['rank'], errors='coerce').astype('Int64')
df['revenue_b$'] = df['revenue_b$'].astype(float)
df['market_cap_b$'] = df['market_cap_b$'].astype(float)

df
```

```python
#VISUALIZATION

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.dpi"] = 120


def rotate(ax, angle=65):
    for lbl in ax.get_xticklabels():
        lbl.set_rotation(angle)
        lbl.set_ha("right")
```

```python
#1. Top companies by revenue

top_n = 20
plt.figure(figsize=(12,5))
sns.barplot(data=df.sort_values("revenue_clean", ascending=False).head(top_n),
            x="organization", y="revenue_clean", color="skyblue")
rotate(plt.gca())
plt.title(f"Top {top_n} Companies by Revenue")
plt.ylabel("Revenue (B$)")
plt.tight_layout()
plt.show()
```

```python
#2. Top companies by market cap

plt.figure(figsize=(12,5))
sns.barplot(data=df.sort_values("market_cap_clean", ascending=False).head(top_n),
            x="organization", y="market_cap_clean", color="orange")
rotate(plt.gca())
plt.title(f"Top {top_n} Companies by Market Cap")
plt.ylabel("Market Cap (B$)")
plt.tight_layout()
plt.show()
```

```python
#3. Revenue vs Market Cap (Scatter)

plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="revenue_clean", y="market_cap_clean", s=70)
plt.title("Revenue vs Market Cap")
plt.xlabel("Revenue (B$)")
plt.ylabel("Market Cap (B$)")
plt.tight_layout()
plt.show()
```

```python
#4. Revenue Distribution (Histogram)

plt.figure(figsize=(8,4))
sns.histplot(df['revenue_clean'], bins=30, kde=True, color="royalblue")
plt.title("Revenue Distribution")
plt.xlabel("Revenue (B$)")
plt.tight_layout()
plt.show()
```

```python
#5. Market Cap Distribution (Histogram)

plt.figure(figsize=(8,4))
sns.histplot(df['market_cap_clean'], bins=30, kde=True, color="green")
plt.title("Market Cap Distribution")
plt.xlabel("Market Cap (B$)")
plt.tight_layout()
plt.show()
```

```python
#6. Revenue Boxplot (Outlier Analysis)

plt.figure(figsize=(6,4))
sns.boxplot(x=df['revenue_clean'], color="skyblue")
plt.title("Revenue Outlier Boxplot")
plt.tight_layout()
plt.show()
```

```python
#8. Compare Revenue & Market Cap (Side-by-side Bar Chart)

plt.figure(figsize=(12,5))
df_plot = df.sort_values("revenue_clean", ascending=False).head(15)
plt.bar(df_plot["organization"], df_plot["revenue_clean"], label="Revenue", alpha=0.7)
plt.bar(df_plot["organization"], df_plot["market_cap_clean"], label="Market Cap", alpha=0.7)
rotate(plt.gca())
plt.legend()
plt.title("Revenue vs Market Cap (Top 15)")
plt.tight_layout()
plt.show()
```
